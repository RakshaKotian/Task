```python
import pandas as pd
import matplotlib.pyplot as plt

#Loading the dataset
data = pd.read_csv("C:/Users/raksh/Downloads/Tipss.csv")

#Bar Chart - Total Bill by Day
plt.figure(figsize=(8, 5))
data.groupby('day')['total_bill'].sum().plot(kind='bar', color='lavender', label='Total Bill by day')
plt.title('BAR GRAPH SHOWING TOTAL BILL BY DAY ')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.legend()
plt.show()
```

```python
#Line Chart - Average Tip by Day
plt.figure(figsize=(8, 5))
data.groupby('day')['tip'].mean().plot(kind='line', marker='*', color='purple', label='Average Tip')
plt.title('LINE PLOT SHOWING AVERAGE TIP BY DAY')
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.legend()
plt.grid(True)
plt.show()
```
